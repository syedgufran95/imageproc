```python
import cv2
```

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
img=cv2.imread(r"C:\Users\MACHINE\Pictures\known_faces\gufran.jpg",cv2.IMREAD_GRAYSCALE)
plt.imshow(img)
plt.show()
img.shape
```

```python
cv2.imwrite("aa.png",img)
```

```python
image_50x50=cv2.resize(img,(80,80))
plt.imshow(image_50x50)
plt.show()
image_50x50.shape
```

```python
cropped=image_50x50[:70,20:60]
plt.imshow(cropped)
plt.show()
cropped.shape
```

```python
image_blur=cv2.blur(cropped,(2,2))
plt.imshow(image_blur)
plt.show()
image_blur.shape
```

```python
kernel=np.ones((5,5))*.1
kernel
image_kernel=cv2.filter2D(img,-1,kernel)
plt.imshow(image_kernel)
plt.show()
plt.imshow(img)
plt.show()
image_kernel.shape
```

```python
kernel=np.array([[0,-1,0],
                [-1,5,-1],
                [0,-1,0]])
image_sharp=cv2.filter2D(image_blur,-1,kernel)
plt.imshow(image_blur)
plt.show()
plt.imshow(image_sharp)
plt.show()
```

```python

enhancedimg=cv2.equalizeHist(img)
plt.imshow(img)
plt.show()
plt.imshow(enhancedimg)
plt.show()
```

```python
img2=cv2.imread(r"C:\Users\MACHINE\Pictures\known_faces\gufran.jpg")
image_yuv=cv2.cvtColor(img2,cv2.COLOR_BGR2YUV)
image_yuv[:,:,0]=cv2.equalizeHist(image_yuv[:,:,0])
img3=cv2.cvtColor(image_yuv,cv2.COLOR_YUV2RGB)
plt.imshow(img2)
plt.show()
plt.imshow(img3)
plt.show()
```
